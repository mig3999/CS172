### Finding Cartesian Products

We can find the Cartesian Product of 2 (or more) sets by using the `cartesian_product` function in Sagemath.

Here's an example:

```
X = Set([1,2,3])
Y = Set([4,5])

for c in cartesian_product([X,Y]):
    print (c)
```

### Finding combinations using `Arrangements()`

SageMath (and Python) have a bunch of functions that can help us with combinatorics.  

For example:

To figure out all the possible combinations of a set, use the `Arrangements()` function.  Here's an example:

````python
# here's a set with the letters KEYANO

A = Set(["K","E","Y","A","N","O"])

print ("The Set A is ...",A)

# we can find the Cardinality of the Set using `len()`

print ("The Cardinality of A is",len(A))

# now let's find the possible arrangements.  
# The `Arrangements` function takes two arguments:  
# First, the Set (can be a variable or the Set itself).  
# Second, the number of elements in permutations.  Be sure to include the parentheses.

print ("The possible permutations of A: ")

# try changing the second argument (after the A,) 
Arrangements(A,3).list()


````
