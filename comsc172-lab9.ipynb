We can create matrices using the `matrix` function, like this:

```python
A = matrix([[1,2],[3,4]])
print (A)
```

Once we have a matrix, we can do the usual matrix math with it.  Here's an example:

```python
A = matrix([[1,2],[3,4]])
B = matrix([[3,5],[7,8]])
print ("A is:")
print (A)
print ("\nB is:")
print (B)
print ("\nA * B:")
print (A * B)
```
