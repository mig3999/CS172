## Truth Tables in Sagemath

In Sagemath, we can create propositional logic from formulas.  For example, let's take a proposition:

"If it's raining, and I don't have an umbrella, then I will get wet."

We can express it like this:

```
w = propcalc.formula("r&~u")
```
Where r = raining, u = umbrella `~` means not.

then we can generate a truth table:

```
w.truthtable()
```

Run the cell below for the result.



````python
w = propcalc.formula("r&~u")
w.truthtable()
````

Notice the "result" column?  If it's raining and you *don't* have an umbrella, you'll get wet.  Duh.

Here are operators you can use in your logic formulas:

* & = and
* | = or
* ~ = not
* ^ = xor (exclusive OR, remember that?)
* -> = if-then
* <-> = if and only if

There's a lot more info here on [the sagemath reference page](https://doc.sagemath.org/pdf/en/reference/logic/logic.pdf)

We can put those together to make a complex compound proposition, using parentheses, etc.
