We can also check for Injectivity, surjectivity, and Bijectivity of functions using solve and range_f methods.

Examples below (try them)

```python
# Check injectivity using 'solve' function 
# to find solutions to f(x) = f(y) and verify 
# that they imply x = y.

solutions = solve(f(x) == f(y), x)
print (solutions)
```

```python
#You can find if a function is surjective by checking if every 
#element in a codomain is present in the domain

# Define the domain and codomain
domain = range(5)  # Domain is {0, 1, 2, 3, 4}
codomain = range(25)  # Codomain is {0, 1, 4, 9, 16, ..., 24}

is_surjective = all(any(f(x) == y for x in domain) for y in codomain)

if is_surjective:
    print("The function is surjective.")
else:
    print("The function is not surjective.")
```
